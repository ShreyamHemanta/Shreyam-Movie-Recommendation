```python
import pandas as pd
```

```python
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# IMPORTANT COLUMNS
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#To find missing data

movies.isnull().sum()
```

```python
#Dropping null columns as they are much less

movies.dropna(inplace=True)
```

```python
#To find duplicate data

movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# Change genres, keywords type from list of dictionaries to list of important words
```

```python
import ast
def convert(obj):
    list = []
    for i in ast.literal_eval(obj): #changing list of string to list
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# Only the 1st 3 dictionaries are used to take the main names of actors
def convert3(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj): #changing list of string to list
        if counter != 3:
            list.append(i['name'])
        else :
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#Fetching names of directors only and not other crew members

def fetch_director(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj): #changing list of string to list
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#Changing overview which is a string into a list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Removing spaces to remove confusing between same first names

movies["genres"] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Concatinating overview, cast, crew and keywords column into tags columns

movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][1300]
```

```python
#Text vectorization

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
